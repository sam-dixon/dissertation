```python
import os
import pickle
import numpy as np
import matplotlib.pyplot as plt
from astropy.cosmology import FlatLambdaCDM, w0waCDM, WMAP9
```

# Cosmology comparisons

```python
z = np.linspace(0.001, 2., 100)
```

```python
sm = WMAP9
no_de = FlatLambdaCDM(H0=70, Om0=1.)
w = w0waCDM(H0=70, Om0=0.3, Ode0=0.7, w0=-0.8, wa=-1)
```

```python
plt.plot(z, sm.luminosity_distance(z).value/1e3,
         label='Fiducial cosmology\n(WMAP9)')
plt.plot(z, no_de.luminosity_distance(z).value/1e3,
         label='Flat $\Lambda$CDM\n$\Omega_\Lambda=0$')
plt.plot(z, w.luminosity_distance(z).value/1e3,
         label='$w_0=-0.8$,\n$w_a=-1$')
plt.legend()
# plt.xscale('log')
plt.xlabel('Redshift')
plt.ylabel('Luminosity distance (GpC)')
plt.savefig('../figures/hubble_diagram_examples.pdf', bbox_inches='tight')
```

```python
plt.plot(z, sm.distmod(z).value - 19.1,
         label='Fiducial cosmology\n(WMAP9)')
plt.plot(z, no_de.distmod(z).value - 19.1,
         label='Flat $\Lambda$CDM\n$\Omega_\Lambda=0$')
plt.plot(z, w.distmod(z).value - 19.1,
         label='$w_0=-0.8$,\n$w_a=-1$')
plt.legend()
plt.xscale('log')
plt.xlabel('Redshift')
plt.ylabel(' (GpC)')
plt.savefig('../figures/hubble_diagram_examples_mB.pdf', bbox_inches='tight')
```
